```python
import io
import requests
import pandas as pd
from urllib.parse import quote
from urllib.parse import urlencode
```

```python
base_url = r'https://olinda.bcb.gov.br/olinda/servico/Expectativas/versao/v1/odata/'
endpoint = r'ExpectativaMercadoMensais'
payload  = {"$format"  : "text/csv",
            "$filter"  : "Indicador eq 'IPCA' and Data gt '2020-01-01'",
            "$orderby" : "DataReferencia"}
```

```python
r = requests.get( 
                  url    = base_url + endpoint, 
                  params = urlencode(payload, quote_via=quote)
                )
```

```python
pd.read_csv(io.StringIO(r.text)).head(30)
```
